```python
import numpy as np
import pandas as pd
```

```python
laptops = pd.read_csv('laptop_price.csv', encoding=('ISO-8859-1'),low_memory=False)
```

```python
laptops['Product'] = laptops['Product'].str.split('(').str[0].str.strip()
```

```python
laptops.iloc[24].Product
```

```python
laptops = laptops[['laptop_ID','Company','Product','Cpu','Ram','Memory','Gpu']]
```

```python
laptops.head(30)
```

```python
laptops.isnull().sum()
```

```python
laptops.duplicated().sum()
```

```python
laptops['cpu_t'] = laptops['Cpu']
laptops['Ram_t'] = laptops['Ram']
laptops['Memory_t'] = laptops['Memory']
laptops['Gpu_t'] = laptops['Gpu']
```

```python
tags_to_modify = ['cpu_t', 'Ram_t', 'Memory_t', 'Gpu_t']
for column in tags_to_modify:
    laptops[column] = laptops[column].str.replace(" ", "")
```

```python
title_to_modify = ['Ram', 'Memory', 'Gpu', 'Cpu']
for column in title_to_modify:
    laptops[column] = laptops[column].str.strip()
```

```python
laptops['tags' ] = laptops['cpu_t'] + ' ' + laptops['Ram_t'] + ' ' + laptops['Memory_t'] + ' ' + laptops['Gpu_t']
laptops['model_name'] = laptops['Company'] + ' ' + laptops['Product']
```

```python
laptops['model_name'] = laptops['model_name'].str.replace('"', '').str.strip()
```

```python
laptops['image_title' ] = laptops['model_name']
laptops['image_title'] = laptops['image_title'].str.replace(' ', '_').str.strip()
```

```python
laptops['laptop_title' ] = laptops['model_name'] + '(' + laptops['Ram'] + '/' + laptops['Memory'] + '/' + laptops['Gpu'] + '/' + laptops['Cpu'] + ')'
```

```python
laptops.head()
```

```python
laptops.iloc[0].laptop_title
```

```python
new_df = laptops[['laptop_ID', 'model_name', 'laptop_title', 'image_title', 'Cpu', 'Ram', 'Memory', 'Gpu', 'tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
new_df.drop_duplicates(subset=["model_name", "Cpu", "Ram", "Memory", "Gpu"], inplace=True)
```

```python
new_df
```

```python
new_df.shape
```

```python
# new_df.to_csv('new_dataframe.csv', index=False)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=500)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(laptop):
    laptop_index = new_df[new_df['laptop_title'] == laptop].index[0]
    distances = similarity[laptop_index]
    laptop_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in laptop_list:
        print(new_df.iloc[i[0]].model_name)
        print(new_df.iloc[i[0]].Cpu)
        print(new_df.iloc[i[0]].Ram)
        print(new_df.iloc[i[0]].Memory)
        print(new_df.iloc[i[0]].Gpu)
```

```python
recommend('Apple MacBook Pro(8GB/128GB SSD/Intel Iris Plus Graphics 640/Intel Core i5 2.3GHz)')
```
